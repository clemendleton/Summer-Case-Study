### Cleaning up the Kansas's County Estimated 2010-2019 population

As we can see Kansas_2010_2019_population_Estimates last few rows are unneeded, we I will remove them  

```python
Kansas_2010_2019_population = Kansas_2010_2019_population_Estimates.iloc[:-5]

Kansas_2010_2019_population.tail(7)
```

Since we have the 2010 population we will remove it from the Kansas_2010_2019_population data. while we would remove the leading dots in the first column. 

```python
#drop first row with kansas state total sum
Kansas_2010_2019 = Kansas_2010_2019_population.drop([0])

#removing leading dots(.)
Kansas_2010_2019["Unnamed: 0"] = Kansas_2010_2019["Unnamed: 0"].str.lstrip('.')
```

As we can notice that the unnameed:0 column have the county and the state in one column sepreated my a comma. we will splite and rename the columns while dropping unwanted columns 

```python
#spliting to have county only 
Kansas_2010_2019[['County_full','State']] = Kansas_2010_2019["Unnamed: 0"].str.split(',', n=1, expand=True)
Kansas_2010_2019[['County','NameCounty']] = Kansas_2010_2019["County_full"].str.split(' ', n=1, expand=True)


#dropping the last columns that do not porvid relitive info
Kansas_2010_2019 = Kansas_2010_2019.drop(Kansas_2010_2019.columns[[0,1,2,3,-1,-3,-4]],axis=1)

#changing column order
cols = Kansas_2010_2019.columns.tolist()
cols = cols[-1:]+cols[:-1]
Kansas_2010_2019 = Kansas_2010_2019[cols]

#Converting nana or '-' to 0
Kansas_2010_2019 = Kansas_2010_2019.fillna(0)
Kansas_2010_2019 = Kansas_2010_2019.replace('-',0)


Kansas_2010_2019.head()
```

we would store the Kansas state 2010-2019 population into a new dataframe

```python
#Kansas State Poplation 
Kansas_population_2010_2019 = Kansas_2010_2019_population[Kansas_2010_2019_population["Unnamed: 0"] == 'Kansas' ]
Kansas_population_2010_2019 = Kansas_population_2010_2019.rename(columns = {'Unnamed: 0' : 'County' }, inplace = False)


#dropping columns that do not porvid relitive info
Kansas_population_2010_2019 = Kansas_population_2010_2019.drop(Kansas_population_2010_2019.columns[[1,2,3]],axis=1)

Kansas_population_2010_2019.head()
```

### Cleaning up the Kansas's County 1860 to 2010 population

Since Kansas_1860_1970_population and Kansas_1980_2010_population hare the same dataformat. We would merge the two datasets into one for easier of use  

```python
#function use to merg years by county
def Merge_Population(x,y,k):
    Kansas_population = x.merge(y, on= k)
    return Kansas_population
```

```python
#calling the Merge_Population function
Kansas_1860_2010_population = Merge_Population(Kansas_1860_1970_population,Kansas_1980_2010_population,'County')

#Converting nana or '-' to 0
Kansas_1860_2010_population = Kansas_1860_2010_population.fillna(0)
Kansas_1860_2010_population = Kansas_1860_2010_population.replace('-',0)

Kansas_1860_2010_population.head()
```

### Converting String/Factors columns to Numeric numbers

Noticing that there are commas in the respected number. instead of leaving them as string/ojects/factors we will convert them into numeric numbers 

```python
#convert County to str & dtype=object to numeric
def covertObejec(df):
    #storing column names and droping the first name to convert the types to numeric
    years = df.columns.tolist()
    years.pop(0)
    #convert and remove the commas
    for year in years:
        df[year] = pd.to_numeric(df[year].str.replace(',', ''),  downcast='integer')
    return df
```

```python
#converting the columns to numeric
Kansas_1860_2019_population = covertObejec(Kansas_1860_2019_population)
Kansas_State_population = covertObejec(Kansas_State_population)
```
