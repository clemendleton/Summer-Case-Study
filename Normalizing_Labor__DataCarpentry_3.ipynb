### Normalizing_Labor_Data

```python
#Import modules
import pandas as pd
```

```python
#reading the souce files
Estimated_Rileypopulation = pd.read_csv('../../../casestudy_data/group_2/Estimated_Riley.population.csv')
Estimated_Pottawatomiepopulation = pd.read_csv('../../../casestudy_data/group_2/Estimated_Pottawatomie.population.csv')
Estimated_Ellispopulation = pd.read_csv('../../../casestudy_data/group_2/Estimated_Ellis.population.csv')
Estimated_Dickinsonpopulation = pd.read_csv('../../../casestudy_data/group_2/Estimated_Dickinson.population.csv')
Kansas_Labor_Data = pd.read_csv('../../../casestudy_data/group_2/Kansas_Unemployment_monthly.csv')
```

#### SubSeting & Merging

###### Since we are choosing only three countieswe are going to subset the counties and add a Employment precentage and Merging the estimated population data together

```python
#subset by county
Riley = Kansas_Labor_Data[Kansas_Labor_Data["County"] == "Riley"]


#sort by time 
Riley = Riley.sort_values(by=["Time"], ascending= True)

#get the Labor_Force change
Riley['Employment'] = Riley['Employed']/Riley['Labor_Force'] * 100

#drop unwanted column
Riley = Riley.drop(Riley.columns[[0]],axis=1)

#merge estimated population data 
Riley_Labor_population = pd.merge(Estimated_Rileypopulation,Riley,how='inner', right_on =['Time','County'], left_on =['Year','County'])  

#drop dulicated year/time column
Riley_Labor_population = Riley_Labor_population.drop(Riley_Labor_population.columns[[3]],axis=1)

#get the Participation Rate
Riley_Labor_population['Labor_Force_Participation_Rate'] = Riley_Labor_population['Labor_Force']/Riley_Labor_population['population'] * 100

#round population by 0 decimal
Riley_Labor_population['population'] = Riley_Labor_population['population'].round(decimals=0)

#change to int 
Riley_Labor_population['population'] = Riley_Labor_population['population'].astype(int)

#round by 1 decimal
Riley_Labor_population['Employment'] = Riley_Labor_population['Employment'].round(decimals=1)
Riley_Labor_population['Labor_Force_Participation_Rate'] = Riley_Labor_population['Labor_Force_Participation_Rate'].round(decimals=1)

Riley_Labor_population.head()

```

```python
#subset by county
Dickinson = Kansas_Labor_Data[Kansas_Labor_Data["County"] == "Dickinson"]

#sort by time 
Dickinson = Dickinson.sort_values(by=["Time"], ascending= True)

#get the Labor_Force change
Dickinson['Employment'] = Dickinson['Employed']/Dickinson['Labor_Force'] * 100

#drop unwanted column
Dickinson = Dickinson.drop(Dickinson.columns[[0]],axis=1)

#merge estimated population data 
Dickinson_Labor_population = pd.merge(Estimated_Dickinsonpopulation,Dickinson,how='inner', right_on =['Time','County'], left_on =['Year','County'])  

#drop dulicated year/time column
Dickinson_Labor_population = Dickinson_Labor_population.drop(Dickinson_Labor_population.columns[[3]],axis=1)

#get the Participation Rate
Dickinson_Labor_population['Labor_Force_Participation_Rate'] = Dickinson_Labor_population['Labor_Force']/Dickinson_Labor_population['population'] * 100

#round population by 0 decimal
Dickinson_Labor_population['population'] = Dickinson_Labor_population['population'].round(decimals=0)

#change to int 
Dickinson_Labor_population['population'] = Dickinson_Labor_population['population'].astype(int)

#round by 1 decimal
Dickinson_Labor_population['Employment'] = Dickinson_Labor_population['Employment'].round(decimals=1)
Dickinson_Labor_population['Labor_Force_Participation_Rate'] = Dickinson_Labor_population['Labor_Force_Participation_Rate'].round(decimals=1)

Dickinson_Labor_population.head()
```

```python
#subset by county
Ellis = Kansas_Labor_Data[Kansas_Labor_Data["County"] == "Ellis"]

#sort by time 
Ellis = Ellis.sort_values(by=["Time"], ascending= True)

#get the Labor_Force change
Ellis['Employment'] = Ellis['Employed']/Ellis['Labor_Force'] * 100

#drop unwanted column
Ellis = Ellis.drop(Ellis.columns[[0]],axis=1)

#merge estimated population data 
Ellis_Labor_population = pd.merge(Estimated_Ellispopulation,Ellis,how='inner', right_on =['Time','County'], left_on =['Year','County'])  

#drop dulicated year/time column
Dickinson_Labor_population = Dickinson_Labor_population.drop(Dickinson_Labor_population.columns[[3]],axis=1)

#get the Participation Rate
Ellis_Labor_population['Labor_Force_Participation_Rate'] = Ellis_Labor_population['Labor_Force']/Ellis_Labor_population['population'] * 100

#round population by 0 decimal
Ellis_Labor_population['population'] = Ellis_Labor_population['population'].round(decimals=0)

#change to int 
Ellis_Labor_population['population'] = Ellis_Labor_population['population'].astype(int)

#round by 1 decimal
Ellis_Labor_population['Employment'] = Ellis_Labor_population['Employment'].round(decimals=1)
Ellis_Labor_population['Labor_Force_Participation_Rate'] = Ellis_Labor_population['Labor_Force_Participation_Rate'].round(decimals=1)

Ellis_Labor_population.head()
```

```python
#subset by county
Pottawatomie = Kansas_Labor_Data[Kansas_Labor_Data["County"] == "Pottawatomie"]

#sort by time 
Pottawatomie = Pottawatomie.sort_values(by=["Time"], ascending= True)

#get the Labor_Force change
Pottawatomie['Employment'] = Pottawatomie['Employed']/Pottawatomie['Labor_Force'] * 100

#drop unwanted column
Pottawatomie = Pottawatomie.drop(Pottawatomie.columns[[0]],axis=1)

#merge estimated population data 
Pottawatomie_Labor_population = pd.merge(Estimated_Pottawatomiepopulation,Pottawatomie,how='inner', right_on =['Time','County'], left_on =['Year','County'])  

#drop dulicated year/time column
Pottawatomie_Labor_population = Pottawatomie_Labor_population.drop(Pottawatomie_Labor_population.columns[[3]],axis=1)

#get the Participation Rate
Pottawatomie_Labor_population['Labor_Force_Participation_Rate'] = Pottawatomie_Labor_population['Labor_Force']/Pottawatomie_Labor_population['population'] * 100

#round population by 0 decimal
Pottawatomie_Labor_population['population'] = Pottawatomie_Labor_population['population'].round(decimals=0)

#change to int 
Pottawatomie_Labor_population['population'] = Pottawatomie_Labor_population['population'].astype(int)

#round by 1 decimal
Pottawatomie_Labor_population['Employment'] = Pottawatomie_Labor_population['Employment'].round(decimals=1)
Pottawatomie_Labor_population['Labor_Force_Participation_Rate'] = Pottawatomie_Labor_population['Labor_Force_Participation_Rate'].round(decimals=1)

Pottawatomie_Labor_population.head()
```
